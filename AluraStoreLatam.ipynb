```python
from geopy.geocoders import Nominatim
import time
import os
```

```python
geolocalizador = Nominatim(user_agent='ventas_geolocalizacion') # Inicializar geolocalizador
```

```python
cache_regiones = {}  # caché global para coordenadas
```

```python
# función para crear el mapa por tienda
def crear_mapa_por_tienda(tienda, nombre_archivo):
  """
    para crear un mapa interactivo con marcadores y un mapa de calor basado en los datos de la tienda.
    """
  mapa = folium.Map(location=[tienda['lat'].mean(), tienda['lon'].mean()], zoom_start = 6) # se crea un mapa base con folium de ubicación promedio de las tiendas

  # creamos una lista - mapa de calor con lat,lon, precio por cada fila del datafrmae
  mapa_de_calor = [[row['lat'], row['lon'], row['Precio']] for index, row in tienda.iterrows()] # iterrows() permite iterar cada fila en el dataframe

  # Esta lista mapa_de_calor alimenta al HeatMap, donde el “peso” (Precio) determina la intensidad del color.
  HeatMap(mapa_de_calor, radius=10, blur=15, max_zoom=1).add_to(mapa) # donde se concentran los precios de los productos.

  # Crea un marcador individual en la ubicación (lat, lon) de la tienda o producto.
  for i, row in tienda.iterrows():
    folium.Marker(location = [row['lat'], row['lon']],
                  popup = f"Producto:{row['Producto']}<br>" # Al hacer clic sobre el marcador, aparece un popup con la información
                            f"Categoria:{row['Categoría del Producto']}<br>"
                        f"Precio: ${row['Precio']:.2f}").add_to(mapa)

  # Guardar el mapa en un archivo HTML
  os.makedirs("docs", exist_ok=True)
  mapa.save(nombre_archivo)
  print(f"Mapa guardado en: {nombre_archivo}")

  return mapa
```

```python
# Funcion para obtener la region
def obtener_region(lat, lon, usar_cache=True):
  """ Obtiene la región (ciudad, estado o país) a partir de coordenadas (lat, lon),
    usando una caché para evitar consultas repetidas al geolocalizador.
    """
  clave = (round(lat, 3), round(lon, 3))  # clave redondeada para evitar duplicados minimos

   # si ya está en caché, se devuelve inmediatamente
  if usar_cache and clave in cache_regiones:
    return cache_regiones[clave]
  try:
    localizacion = geolocalizador.reverse((lat,lon), language='es')
    if localizacion and 'address' in localizacion.raw:
      direccion = localizacion.raw['address']
      region = direccion.get('city', direccion.get('state', direccion.get('country', 'Desconocido')))
    else:
      return 'Desconocido'
  except Exception as e:
        return f'Error: {e}'
  # Guardamos resultado en caché
  if usar_cache:
    cache_regiones[clave] = region

  return region
```

```python
# Función para analizar zonas y obtener top ventas
def analizar_zona_tienda_precio(tienda, obtener_region, esperar=True, delay=1, usar_cache=True ):
  """
    Agrupa las ventas por coordenadas aproximadas, obtenemos la región y devuelve las zonas con más ventas.
    Usamos caché para evitar consultas repetidas al geolocalizador.
    """
  tienda['Latitud Aproximada'] = tienda['lat'].round(1) # Redondeamos coordenadas para agrupar por "zona"
  tienda['Longitud Aproximada'] = tienda['lon'].round(1)
  zonas = tienda.groupby(['Latitud Aproximada', 'Longitud Aproximada']).agg({'Precio': 'sum'}).reset_index()

  # Obtener regiones
  regiones = []
  for _, row in zonas.iterrows():
    region = obtener_region(row['Latitud Aproximada'],row['Longitud Aproximada'],usar_cache = usar_cache)
    regiones.append(region)
    if esperar:
      time.sleep(delay)

  zonas['Region'] = regiones

  zonas_mayor_ventas = zonas.sort_values('Precio', ascending=False)

  print(f"Se analizaron {len(zonas)} zonas.")
  print(f"Caché actual: {len(cache_regiones)} coordenadas guardadas.")

  return zonas_mayor_ventas
```

```python
# Función para analizar zonas y obtener el numero de top de ventas que deseemos
def analizar_zona_tienda(tienda, obtener_region, top_n, esperar=True, delay=1, usar_cache=True ):
  """
    Agrupa las ventas por coordenadas aproximadas, obtiene la región y devuelve las zonas con más ventas.
    Usa caché para evitar consultas repetidas al geolocalizador.
    """
  tienda['Latitud Aproximada'] = tienda['lat'].round(1) # Redondeamos coordenadas para agrupar por "zona"
  tienda['Longitud Aproximada'] = tienda['lon'].round(1)
  zonas = tienda.groupby(['Latitud Aproximada', 'Longitud Aproximada']).agg({'Precio': 'sum'}).reset_index()

  # Obtener regiones
  regiones = []
  for _, row in zonas.iterrows():
    region = obtener_region(row['Latitud Aproximada'],row['Longitud Aproximada'],usar_cache = usar_cache)
    regiones.append(region)
    if esperar:
      time.sleep(delay)

  zonas['Region'] = regiones

  zonas_mayor_ventas = zonas.sort_values('Precio', ascending=False).head(top_n)

  print(f"Se analizaron {len(zonas)} zonas. Mostrando las {top_n} con mayores ventas.")
  print(f"Caché actual: {len(cache_regiones)} coordenadas guardadas.")

  return zonas_mayor_ventas

```

```python
# función para crear el mapa y usarla con la funcion analizar_y_mapear_tienda()
def crear_mapa_por_tienda_top_n(tienda, nombre_archivo):
  """
    Crea un mapa interactivo con marcadores y un mapa de calor basado en los datos de una tienda.
    """
  mapa = folium.Map(location=[tienda['Latitud Aproximada'].mean(), tienda['Longitud Aproximada'].mean()], zoom_start = 6) # se crea un mapa base con folium de ubicación promedio de las tiendas

  # creamos una lista - mapa de calor con lat,lon, precio por cada fila del datafrmae
  mapa_de_calor = [[row['Latitud Aproximada'], row['Longitud Aproximada'], row['Precio']] for index, row in tienda.iterrows()] # iterrows() permite iterar cada fila en el dataframe

  # Esta lista mapa_de_calor alimenta al HeatMap, donde el “peso” (Precio) determina la intensidad del color.
  HeatMap(mapa_de_calor, radius=10, blur=15, max_zoom=1).add_to(mapa) # donde se concentran los precios de los productos.

  # Crea un marcador individual en la ubicación (lat, lon) de la tienda o producto.
  for i, row in tienda.iterrows():
    folium.Marker(location = [row['Latitud Aproximada'], row['Longitud Aproximada']],
                  popup = f"Precio: ${row['Precio']:.2f}").add_to(mapa)  # Al hacer clic sobre el marcador, aparece un popup con la información

  # Guardar el mapa en un archivo HTML
  mapa.save(nombre_archivo)
  print(f"Mapa guardado en: {nombre_archivo}")

  return mapa
```

```python
def analizar_y_mapear_tienda(df_tienda, nombre_archivo, top_n, esperar=False, delay=1,
                             usar_cache=True, incluir_heatmap=True):
  """
    Combina el análisis de zonas con la creación del mapa de ventas.
    Retorna el DataFrame de zonas con más ventas y genera el mapa HTML de esas zonas.
    """
  zonas_top = analizar_zona_tienda(df_tienda, obtener_region, top_n)

  mapa = crear_mapa_por_tienda_top_n(zonas_top, nombre_archivo)

  return zonas_top, mapa
```

## Graficos

```python
import numpy as np
import plotly.express as px

# Copiar dataframe
df_plot = zonas_top5_tiendas_concatenadas.copy()

# Añadir jitter
jitter_strength = 0.01
df_plot["Latitud"] = df_plot["Latitud Aproximada"] + np.random.uniform(-jitter_strength, jitter_strength, size=len(df_plot))
df_plot["Longitud"] = df_plot["Longitud Aproximada"] + np.random.uniform(-jitter_strength, jitter_strength, size=len(df_plot))

# Crear gráfico con paleta Viridis
fig = px.scatter(
    df_plot,
    x="Longitud",
    y="Latitud",
    size="Precio",
    color="Tienda",
    hover_name="Region",
    hover_data={
        "Precio": True,
        "Region": True,
        "Latitud Aproximada": True,
        "Longitud Aproximada": True
    },
    size_max=50,
    opacity=0.8,
    title="Mapa de ventas por ubicación y tienda (con puntos diferenciados)",
    facet_col="Tienda",
    facet_col_wrap=2,
    color_discrete_sequence=px.colors.sequential.Viridis  # Paleta Viridis
)

# Personalizar ejes globales
fig.update_layout(
    legend_title="Tienda",
    height=600
)

fig.show()
```
