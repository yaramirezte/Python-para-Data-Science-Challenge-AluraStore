# 4. Productos más y menos vendidos

```python
#Los top 5 mas vendido por tienda
df_resultados_top5 =[]
for i, x in enumerate([tienda1,tienda2,tienda3,tienda4], start=1):
  top5 =x['Producto'].value_counts().head(5)
  for producto, cantidad in top5.items():
    df_resultados_top5.append({'Tienda': f'Tienda{i}',
                       'Producto': producto,
                       'Cantidad Vendida': cantidad})
Ventas_por_producto_top5 = pd.DataFrame(df_resultados_top5)
Ventas_por_producto_top5
```

```python
# se genera este codigo graficar y tener el mas vendido por cada tienda
df_resultados =[]
for i, x in enumerate([tienda1,tienda2,tienda3,tienda4], start=1):
  top5 =x['Producto'].value_counts().head(1)
  for producto, cantidad in top5.items():
    df_resultados.append({'Tienda': f'Tienda{i}',
                       'Producto': producto,
                       'Cantidad Vendida': cantidad})
Ventas_por_producto = pd.DataFrame(df_resultados)
Ventas_por_producto
```

```python
# Bottom 5 menos vendidos por tienda
df_resultados1_bottom5 =[]
for i, x in enumerate([tienda1,tienda2,tienda3,tienda4], start=1):
  bottom =x['Producto'].value_counts().tail(5)
  for producto, cantidad in bottom.items():
    df_resultados1_bottom5.append({'Tienda': f'Tienda{i}',
                       'Producto': producto,
                       'Cantidad Vendida': cantidad})
menos_ventas_por_producto_bottom5 = pd.DataFrame(df_resultados1_bottom5)
menos_ventas_por_producto_bottom5
```

```python
## se genera este codigo graficar y tener el menos vendido por cada tienda
df_resultados1 =[]
for i, x in enumerate([tienda1,tienda2,tienda3,tienda4], start=1):
  bottom =x['Producto'].value_counts().tail(1)
  for producto, cantidad in bottom.items():
    df_resultados1.append({'Tienda': f'Tienda{i}',
                       'Producto': producto,
                       'Cantidad Vendida': cantidad})
menos_ventas_por_producto = pd.DataFrame(df_resultados1)
menos_ventas_por_producto
```

```python
import numpy as np

Ventas_por_producto['Tipo'] = 'Más vendido'
menos_ventas_por_producto['Tipo'] = 'Menos vendido'

comparacion_ventas = pd.concat([Ventas_por_producto, menos_ventas_por_producto])# unimos ambos
comparacion_ventas
```

```python
import matplotlib.pyplot as plt

# Suponiendo que ya tienes el DataFrame 'comparacion_ventas'
tiendas = comparacion_ventas['Tienda'].unique()
tipos = comparacion_ventas['Tipo'].unique()
x = np.arange(len(tiendas))

width = 0.35  # ancho de barra
plt.figure(figsize=(10,6))

colores = ['#bddf26', '#29af7f']  # color para cada tipo

for i, tipo in enumerate(tipos):
    subset = comparacion_ventas[comparacion_ventas['Tipo'] == tipo]
    plt.bar(x + i*width, subset['Cantidad Vendida'], width=width, label=tipo, color=colores[i])

    # Etiquetas de producto encima de cada barra
    for j, v in enumerate(subset['Cantidad Vendida']):
        producto = subset.iloc[j]['Producto']
        plt.text(
            x[j] + i*width, v + 1, producto,
            ha='center', rotation=20, fontsize=8,  # rotación leve
            rotation_mode='anchor'
        )

plt.title('Producto Más y Menos Vendido por Tienda', pad=25)  # separa el título del gráfico
plt.xlabel('Tienda')
plt.ylabel('Cantidad Vendida')
plt.xticks(x + width/2, tiendas)
plt.legend(title='Tipo de producto')

plt.subplots_adjust(top=0.88)  #  da más espacio arriba
plt.tight_layout()
plt.show()
```
