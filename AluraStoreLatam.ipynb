```python
resultados =[]
for i, x in enumerate([tienda1,tienda2,tienda3,tienda4], start=1):
  top5 =x['Categoría del Producto'].value_counts()
  for categoria, cantidad in top5.items():
    resultados.append({'Tienda': f'Tienda{i}',
                       'Categoria': categoria,
                       'Cantidad Vendida': cantidad})
Ventas_por_categoria = pd.DataFrame(resultados)
Ventas_por_categoria
```

```python
suma_por_categoria = Ventas_por_categoria.groupby('Categoria')['Cantidad Vendida'].sum().sort_values()
suma_por_categoria
```

# 4. Productos más y menos vendidos

```python
#Los top 5 mas vendido por tienda
df_resultados_top5 =[]
for i, x in enumerate([tienda1,tienda2,tienda3,tienda4], start=1):
  top5 =x['Producto'].value_counts().head(5)
  for producto, cantidad in top5.items():
    df_resultados_top5.append({'Tienda': f'Tienda{i}',
                       'Producto': producto,
                       'Cantidad Vendida': cantidad})
Ventas_por_producto_top5 = pd.DataFrame(df_resultados_top5)
Ventas_por_producto_top5
```

```python
# se genera este codigo graficar y tener el mas vendido por cada tienda
df_resultados =[]
for i, x in enumerate([tienda1,tienda2,tienda3,tienda4], start=1):
  top5 =x['Producto'].value_counts().head(1)
  for producto, cantidad in top5.items():
    df_resultados.append({'Tienda': f'Tienda{i}',
                       'Producto': producto,
                       'Cantidad Vendida': cantidad})
Ventas_por_producto = pd.DataFrame(df_resultados)
Ventas_por_producto
```

```python
# Bottom 5 menos vendidos por tienda
df_resultados1_bottom5 =[]
for i, x in enumerate([tienda1,tienda2,tienda3,tienda4], start=1):
  bottom =x['Producto'].value_counts().tail(5)
  for producto, cantidad in bottom.items():
    df_resultados1_bottom5.append({'Tienda': f'Tienda{i}',
                       'Producto': producto,
                       'Cantidad Vendida': cantidad})
menos_ventas_por_producto_bottom5 = pd.DataFrame(df_resultados1_bottom5)
menos_ventas_por_producto_bottom5
```

```python
## se genera este codigo graficar y tener el menos vendido por cada tienda
df_resultados1 =[]
for i, x in enumerate([tienda1,tienda2,tienda3,tienda4], start=1):
  bottom =x['Producto'].value_counts().tail(1)
  for producto, cantidad in bottom.items():
    df_resultados1.append({'Tienda': f'Tienda{i}',
                       'Producto': producto,
                       'Cantidad Vendida': cantidad})
menos_ventas_por_producto = pd.DataFrame(df_resultados1)
menos_ventas_por_producto
```

```python
import numpy as np

Ventas_por_producto['Tipo'] = 'Más vendido'
menos_ventas_por_producto['Tipo'] = 'Menos vendido'

comparacion_ventas = pd.concat([Ventas_por_producto, menos_ventas_por_producto])# unimos ambos
comparacion_ventas
```

## Ingresos totales por tienda

   El analisis de ingresos muestra que la Tienda1 obtuvo los mayores ingresos totales,seguida por la Tienda2. Las Tiendas3 y Tienda4 presentan ingresos significativamente menores.

```python
plt.figure(figsize=(8,5))
plt.bar(ingresos_tiendas['Tiendas'], ingresos_tiendas['Ingresos Totales'].astype(float), color='skyblue')
plt.title('Ingresos Totales por Tienda')
plt.xlabel('Tiendas')
plt.ylabel('Ingresos ($)')
# Etiquetas encima de cada barra
for i, v in enumerate(ingresos_tiendas['Ingresos Totales']):
    #plt.text(i, v + max(ingresos_tiendas['Ingresos Totales'])*0.01, str(v), ha='center', fontsize=9, fontweight='bold')
    plt.text(i, v + max(ingresos_tiendas['Ingresos Totales'])*0.01, f"${v:,.2f}", ha='center', fontweight='bold', fontsize=9)
plt.tight_layout()
plt.show()
```

## El productos mas y el menos vendido por tienda

```python
tiendas = comparacion_ventas['Tienda'].unique()
tipos = comparacion_ventas['Tipo'].unique()
x= np.arange(len(tiendas))

width= 0.35 #el ancho de barra
plt.figure(figsize=(10,6))

colores= ['skyblue', 'salmon']# color para cada tipo

for i, tipo in enumerate(tipos):
  subset = comparacion_ventas[comparacion_ventas['Tipo'] == tipo]
  plt.bar(x+i*width, subset['Cantidad Vendida'], width= width, label= tipo, color=colores[i])

for j, v in enumerate(subset['Cantidad Vendida']):  # Etiquetas de producto encima de cada barra
  producto= subset.iloc[j]['Producto']
  plt.text(x[j]+ i*width, v+0.1, producto, ha='center', rotation=90, fontsize=8)

plt.title('Producto Más y Menos Vendido por Tienda')
plt.xlabel('Tienda')
plt.ylabel('Cantidad Vendida')
plt.xticks(x + width/2, tiendas)
plt.legend(title='Tipo de producto')
plt.tight_layout()
plt.show()
```
