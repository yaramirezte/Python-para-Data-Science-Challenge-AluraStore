```python
print(f"Ingreso total de la tienda 1: {ingreso_total_tienda1}")
print(f"Ingreso total de la tienda 2: {ingreso_total_tienda2}")
print(f"Ingreso total de la tienda 3: {ingreso_total_tienda3}")
print(f"Ingreso total de la tienda 4: {ingreso_total_tienda4}")
```

# 2. Ventas por categoría

```python
resultados =[]
for i, x in enumerate([tienda1,tienda2,tienda3,tienda4], start=1):
  top5 =x['Categoría del Producto'].value_counts().head(5)
  for categoria, cantidad in top5.items():
    resultados.append({'Tienda': f'Tienda{i}',
                       'Categoria': categoria,
                       'Cantidad Vendida': cantidad})
Ventas_por_categoria = pd.DataFrame(resultados)
Ventas_por_categoria
```
