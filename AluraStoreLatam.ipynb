```python
resultados =[]
for i, x in enumerate([tienda1,tienda2,tienda3,tienda4], start=1):
  top5 =x['Categoría del Producto'].value_counts()
  for categoria, cantidad in top5.items():
    resultados.append({'Tienda': f'Tienda{i}',
                       'Categoria': categoria,
                       'Cantidad Vendida': cantidad})
Ventas_por_categoria = pd.DataFrame(resultados)
Ventas_por_categoria
```

## Ingresos totales de las tiendas

```python
plt.figure(figsize=(8,5))
plt.bar(ingresos_tiendas['Tiendas'], ingresos_tiendas['Ingresos Totales'].astype(float), color='skyblue')
plt.title('Ingresos Totales por Tienda')
plt.xlabel('Tiendas')
plt.ylabel('Ingresos ($)')
# Etiquetas encima de cada barra
for i, v in enumerate(ingresos_tiendas['Ingresos Totales']):
    #plt.text(i, v + max(ingresos_tiendas['Ingresos Totales'])*0.01, str(v), ha='center', fontsize=9, fontweight='bold')
    plt.text(i, v + max(ingresos_tiendas['Ingresos Totales'])*0.01, f"${v:,.2f}", ha='center', fontweight='bold', fontsize=9)
plt.tight_layout()
plt.show()
```

## Valoración media por tienda

```python
# Ordenar de mayor a menor
promedios_ordenado = promedios.sort_values(by='Calificación promedio', ascending=False)

# Gráfico horizontal
plt.figure(figsize=(8,5))
plt.barh(promedios_ordenado['Tienda'], promedios_ordenado['Calificación promedio'], color='lightgreen')
plt.xlabel('Calificación Promedio')
plt.ylabel('Tienda')
plt.xlim(0,5)

# Etiquetas encima de cada barra
for i, v in enumerate(promedios_ordenado['Calificación promedio']):
    plt.text(v + 0.05, i, str(v), va='center', fontweight='bold')

plt.title('Calificación Promedio por Tienda (Mayor a Menor)')
plt.tight_layout()
plt.show()
```
